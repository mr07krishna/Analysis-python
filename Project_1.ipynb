**Customer Churn Analysis and Visualization**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DATA/Customer Churn.csv')
```

```python
print(df.head)
```

```python
print(df.info())
```

**Key Visualizations**

1. Overall Churn Rate

```python
plt.figure(figsize=(6,6))
df['Churn'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#66b3ff','#ff9999'])
plt.title('Customer Churn Distribution')
plt.ylabel('')
plt.show()
```

2. Churn by Contract Type

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Contract', hue='Churn', data=df, palette=['#66b3ff','#ff9999'])
plt.title('Churn by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.show()
```

3. Churn by Internet Service Type

```python
plt.figure(figsize=(10,6))
sns.countplot(x='InternetService', hue='Churn', data=df, palette=['#66b3ff','#ff9999'])
plt.title('Churn by Internet Service Type')
plt.xlabel('Internet Service')
plt.ylabel('Number of Customers')
plt.show()
```

4. Monthly Charges vs. Churn

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Churn', y='MonthlyCharges', data=df, palette=['#66b3ff','#ff9999'])
plt.title('Monthly Charges Distribution by Churn Status')
plt.xlabel('Churn Status')
plt.ylabel('Monthly Charges ($)')
plt.show()
```

5. Tenure Distribution by Churn Status

```python
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', palette=['#66b3ff','#ff9999'])
plt.title('Tenure Distribution by Churn Status')
plt.xlabel('Tenure (months)')
plt.ylabel('Number of Customers')
plt.show()
```

6. Payment Method Impact on Churn

```python
plt.figure(figsize=(12,6))
sns.countplot(x='PaymentMethod', hue='Churn', data=df, palette=['#66b3ff','#ff9999'])
plt.title('Churn by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

**7. Additional Services Impact on Churn**

```python
services = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

plt.figure(figsize=(15,10))
for i, service in enumerate(services, 1):
    plt.subplot(2,3,i)
    sns.countplot(x=service, hue='Churn', data=df, palette=['#66b3ff','#ff9999'])
    plt.title(f'Churn by {service}')
    plt.xlabel(service)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show(
)
```
